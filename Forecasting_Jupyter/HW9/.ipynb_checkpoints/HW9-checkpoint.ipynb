```python
import pandas as pd
import statsmodels as sm
from statsmodels import api as smi
from statsmodels import graphics as smg
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import math
import itertools as it
from arch import arch_model
from arch.univariate import ARCH
from arch.univariate import GARCH

%matplotlib inline
```

```python
# get data
dat = pd.read_table('Long.M', header=None, names=['Long'])
dat['Short'] = pd.read_table('Short.M', header=None, names=['Short'])
dat['Spread'] = dat['Long'] - dat['Short']
dat['LagSpread'] = dat['Spread'].shift()
dat['const'] = 1
dat['diffSpread'] = dat['Spread'] - dat['LagSpread']
dat
```

```python
#plot spread

fig, SPRplot = plt.subplots(nrows=1, ncols=1, figsize=(14,5))
legLine0 = mlines.Line2D([],[],color='b',markersize=15, label='Chaos 1') 
legLine1 = mlines.Line2D([],[],color='m',markersize=15, label='Chaos 2')

SPRplot.set_title('Spread')
SPRplot.plot(dat['Spread'],  color='b', linestyle='-', linewidth=2)
```

```python
# Spread ACF, PACF
smg.tsaplots.plot_acf(dat['Spread'], zero=False, lags=50)
plt.title('Spread ACF')
smg.tsaplots.plot_pacf(dat['Spread'],zero=False, lags=50)
plt.title('Spread PACF')
```

```python
# WITH LAG SPREAD AS DEPENDANT VARIABLE 

model = sm.regression.linear_model.OLS(dat['LagSpread'][1:], dat.loc[1:, ['const','Spread']],  hasconst=True)
results1 = model.fit()

coef = results1.params[1]
SE = results1.bse['Spread']
Pval = (coef - 1)/SE
```

```python
# WITH SPREAD AS DEPENDANT VARIABLE 

model = sm.regression.linear_model.OLS(dat['Spread'][1:], dat.loc[1:, ['const','LagSpread']],  hasconst=True)
results1 = model.fit()

coef = results1.params[1]
SE = results1.bse['LagSpread']
Pval = (coef - 1)/SE
```

```python
# T-table critical value is 1 tail 2.364 and 2 tail 1.660 
# D-W T table critical value is -3.51 at .01 with 100 DF 
print("Coef = %#f\nSE = %#f\nPval = %#f" %(coef, SE, Pval))
```
