```python
import requests
import json
import datetime
from datetime import timedelta, date
import time
import pandas as pd
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)

def getWeatherDat(startDate, endDate):
    
    dateList = []
    #start_date = date(2015, 1, 1)
    #end_date = datetime.date.today()
    start_date = startDate
    end_date = endDate
    for single_date in daterange(start_date, end_date):
        dateList.append((single_date,int(time.mktime(single_date.timetuple()))))

    DSapiKey = 'c92c79203c646d6b3e5d6dc3c6381cf8'
    latLong = '40.730436, -73.997461'
    excludBlocks = 'currently,minutely,hourly,alerts,flags'
    
    results = []
    
    for date in dateList: 
        reqTime = date[1]
        SOURCE = 'https://api.darksky.net/forecast/%s/%s,%s?exclude=%s' %(DSapiKey,latLong,reqTime,excludBlocks)

        response = requests.get(SOURCE)
        responseTxt = response.text
        respJson = json.loads(responseTxt)
        
        res = [date[0].isoformat(), respJson['daily']['data']]
        results.append(res)

    
    return results 
```

```python
weaDat_all = getWeatherDat(date(2015, 1, 1), datetime.date.today())
```

```python
# create DataFrame from queried weather data 

forDF = []

for day in weaDat_all:
    date = day[0]
    highTemp = day[1][0]['apparentTemperatureHigh']
    precipIntensity = day[1][0]['precipIntensity']
    precipProb = day[1][0]['precipProbability']
    
    forDF.append([date, highTemp,precipIntensity,precipProb])
    
wea_datDF = pd.DataFrame(forDF, columns = ['date', 'apparentTemperatureHigh','precipIntensity', 'precipProbability'])
```
