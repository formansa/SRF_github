```python
from statsmodels import regression as SMRG
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import pygal as pg
%matplotlib inline
```

```python
bk3dat = pd.read_csv('BK3_tonnage_dat.csv')
bk3dat['Mo'] = bk3dat['MONTH'][-3:]
bk3dat.set_index('MONTH',inplace=True)

#for i in bk3dat['MONTH']: print(i[-3:])
bk3dat
```

```python
# three time series charts
bk3dat.iloc[:,2:5].plot(subplots=True, figsize=(6, 6))
```

```python
# three time series charts on the same axis 
ylims = [bk3dat.iloc[:,3:5].min().min() - 20 ,bk3dat.iloc[:,3:5].max().max() + 20]

fig, axes = plt.subplots(nrows = 1, ncols=3)
bk3dat['REFUSETONSCOLLECTED'].plot(ax=axes[0],figsize=(14,3)) 
axes[0].set_title('REFUSETONSCOLLECTED')
#axes[0].set_ylim(ylims)

bk3dat['PAPERTONSCOLLECTED'].plot(ax=axes[1],figsize=(14,3))
axes[1].set_title('PAPERTONSCOLLECTED')
axes[1].set_ylim(ylims)

bk3dat['MGPTONSCOLLECTED'].plot(ax=axes[2],figsize=(14,3))
axes[2].set_title('MGPTONSCOLLECTED')
axes[2].set_ylim(ylims)
```

```python
# Scatter matrix! needs improvement to be handed in. 
pd.scatter_matrix(bk3dat.iloc[:,2:5], alpha=0.2, figsize=(6, 6), diagonal='kde')
#pd.scatter_matrix(bk3dat.iloc[:,3:6], alpha=0.2, figsize=(6, 6), grid=False)
#pd.scatter_matrix()
```

```python
model = SMRG.linear_model.OLS(bk3dat.iloc[:,2],bk3dat.iloc[:,3:5])
res = model.fit()
print(res.summary2())
```

```python
#residual plots 
stats.probplot(res.resid, plot=plt)
```

```python
resid_hist = res.resid.hist()
resid_hist.set_title('Residual Histogram')

resid_hist
```
