```python
import pandas as pd
import statsmodels as sm
from statsmodels import api as smi
from statsmodels import graphics as smg
from statsmodels import tsa as tsa 
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
%matplotlib inline
```

```python
def tent_map(x):
    '''function defining the generation of chaos numbers'''
    #if 0 <= x <= .6:
    if x >= 0 and x<= .6: 
        res = x/.6
    #if .6 < x <= 1:
    elif x > .6 and x <= 1:
        res = (1-x)/.4
    else:
        res = 'err'
        
    return res
```

```python
chaos1  =  pd.read_csv('CHAOS1.txt', header=0, names=['Chaos1'])
chaos2  =  pd.read_csv('CHAOS2.txt', header=0, names=['Chaos2'])
chaosDat = pd.concat([chaos1, chaos2], axis=1)
```

```python
chaosDat['CH1_check'] = chaosDat['Chaos1'].map(tent_map)
```

```python
chaosDat
```

```python
#plot chaos 1 & 2

fig, chaosPlot = plt.subplots(nrows=1, ncols=2, figsize=(14,5))
legLine0 = mlines.Line2D([],[],color='b',markersize=15, label='Chaos 1') 
legLine1 = mlines.Line2D([],[],color='m',markersize=15, label='Chaos 2')

chaosPlot[0].set_title('Chaos 1')
#chaosPlot[0].set_xlabel('TIME')
#chaosPlot[0].set_ylabel('Russell')
chaosPlot[1].set_title('Chaos 2')
#chaosPlot[1].set_xlabel('TIME')
#chaosPlot[1].set_ylabel('Russell')
chaosPlot[0].plot(chaosDat['Chaos1'], color='b', linestyle='-', marker='o', linewidth=1)
chaosPlot[1].plot(chaosDat['Chaos2'],  color='m', linestyle='-', marker='o', linewidth=1)
# chaosPlot[0].grid()
chaosPlot[0].legend(handles=[legLine0])
chaosPlot[1].legend(handles=[legLine1])
```

```python
# plot ACF and PACF for Chaos 1
smg.tsaplots.plot_acf(chaosDat['Chaos1'], zero=False)
plt.title('Chaos1 ACF')

smg.tsaplots.plot_pacf(chaosDat['Chaos1'], zero=False)
plt.title('Chaos1 PACF')
```

```python
#plot chaos 1 & 2

fig, chaosPlot = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
legLine0 = mlines.Line2D([],[],color='b',markersize=15, label='Chaos 1') 
legLine1 = mlines.Line2D([],[],color='m',markersize=15, label='Chaos 2')

chaosPlot.set_title('Chaos 1 & 2')
chaosPlot.plot(chaosDat['Chaos1'], color='b', linestyle='-', marker='o', linewidth=1)
chaosPlot.plot(chaosDat['Chaos2'],  color='m', linestyle='-', marker='o', linewidth=1)
chaosPlot.legend(handles=[legLine0, legLine1])
```
