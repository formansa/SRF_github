```python
import statsmodels.api as sm
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import pygal as pg
import math
%matplotlib inline
```

```python
GDPdat_initial =  pd.read_csv('GDP.csv')
GDPdat_initial['LogGDP'] = GDPdat_initial['GDP'].apply(math.log)
GDPdat_initial.set_index('DATE', inplace=True)
# insert time predictor variable, intergers 1 to end
GDPdat_initial.insert(2,'TIME',list(range(1,len(GDPdat_initial['LogGDP'])+1)))
GDPdat_initial.head(10)
```

```python
# OLS regression

Y_var = GDPdat_initial.iloc[:,1]
X_var = GDPdat_initial.iloc[:,2]
# add constant manually (as separate step) in statsmodels. Why???
X_var = sm.add_constant(X_var)
model = sm.OLS(Y_var,X_var)
results = model.fit()
print(results.summary())
```

```python
# by default returns 95% CI for coefficients
results.conf_int()
```

```python
# b/c constant added to ols preiction must have constant too
# not sure why, and not sure why 1 works but the result matches minitab... 
# 95% CI on  prediction is sum of standard errors *2 (to get to 2 std dev out) above and below prediciton 
pred_fit = results.predict([1,285])
target_SE = results.bse['TIME'] + results.bse['const']

print('Forecast log(GDP), Q4 2017:\nPoint Forecast: %.5f; 95pct CI: %.5f, %.5f.' %(pred_fit, pred_fit - 2*target_SE, pred_fit + 2*target_SE)) 
```

```python
results.tvalues(0.025)
```

```python
fig, axes = plt.subplots(nrows = 1, ncols=2, figsize=(15, 10))
legend_line = mlines.Line2D([],[],color='red',markersize=15, label='OLS Fitted Line')

GDPdat_initial.iloc[:,0].plot(ax=axes[0], subplots=True)
axes[0].set_title('GDP vs. Time')


GDPdat_initial.iloc[:,1].plot(ax=axes[1], subplots=True)
plt.plot(GDPdat_initial['TIME'], results.fittedvalues.values, color='r', linestyle='-', linewidth=2)
plt.legend(handles=[legend_line])
axes[1].set_title('Log(GDP) vs. Time')
```

```python
#plot of residuals vs time 

fig, resid_plot = plt.subplots()
resid_plot.set_title('Residuals vs. TIME')
resid_plot.set_xlabel('TIME')
resid_plot.set_ylabel('Residual')
resid_plot.plot(GDPdat_initial['TIME'], np.zeros((283,)), color='r', linestyle='-', linewidth=1)
resid_plot.plot(results.resid.values, '-o', mfc='orange')
resid_plot.grid()
```
