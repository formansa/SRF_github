```python
import statsmodels.api as sm
from statsmodels import graphics as smg
import numpy as np
from scipy import stats as SPstats
from time import strptime
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
# import pygal as pg
import math
%matplotlib inline
```

```python
# import csv
RUSSELLdat_initial =  pd.read_csv('Russell.csv')
RUSSELLdat_initial.set_index('Date', inplace=True)
```

```python
# create dataframe 

# create mean-adjusted russell 
AvgRus = np.mean(RUSSELLdat_initial['Close'])
RUSSELLdat_initial['MA_Russ'] = RUSSELLdat_initial['Close'] - AvgRus

# lag russell close 1 and add to DF
YesterdayRuss = RUSSELLdat_initial['MA_Russ'].shift(1)
RUSSELLdat_initial['YesterdayRuss'] = YesterdayRuss[:]
RUSSELLdat_initial['Half_YesterdayRuss'] = .5*RUSSELLdat_initial['YesterdayRuss']
closeLag1 = RUSSELLdat_initial['Close'].shift(1)
RUSSELLdat_initial['Returns'] = (RUSSELLdat_initial['Close'] - closeLag1) / closeLag1
RUSSELLdat_initial['LagReturns'] = RUSSELLdat_initial['Returns'].shift(1)
RUSSELLdat_initial.insert(0,'TIME',list(range(1,len(RUSSELLdat_initial['Open'])+1)))

```

```python
RusselDat = RUSSELLdat_initial
print(RusselDat.shape)
RusselDat 
```

```python
#plot of russel vs time 

fig, russ_plot = plt.subplots()
legend_line = mlines.Line2D([],[],color='b',markersize=15, label='Mean-Adj Russ.') 
legLine2 = mlines.Line2D([],[],color='r',markersize=15, label='Yesterday Russ.')
russ_plot.set_title('Russell vs. TIME')
russ_plot.set_xlabel('TIME')
russ_plot.set_ylabel('Russell')
russ_plot.plot(RusselDat['TIME'], RusselDat['MA_Russ'], color='b', linestyle='-', linewidth=1)
russ_plot.plot(RusselDat['TIME'], RusselDat['YesterdayRuss'],  color='r', linestyle='-', linewidth=1)
russ_plot.grid()
russ_plot.legend(handles=[legend_line, legLine2])


```

```python
#plot of russel vs time 

fig, russ_plot = plt.subplots()
legend_line = mlines.Line2D([],[],color='b',markersize=15, label='Mean-Adj Russ.') 
legLine2 = mlines.Line2D([],[],color='m',markersize=15, label='.5*Yesterday Russ.')
russ_plot.set_title('Russell vs. TIME')
russ_plot.set_xlabel('TIME')
russ_plot.set_ylabel('Russell')
russ_plot.plot(RusselDat['TIME'], RusselDat['MA_Russ'], color='b', linestyle='-', linewidth=1)
russ_plot.plot(RusselDat['TIME'], RusselDat['Half_YesterdayRuss'],  color='m', linestyle='-', linewidth=1)
russ_plot.grid()
russ_plot.legend(handles=[legend_line, legLine2])
```

```python
# Calculate MSE for both time series by fitting the time series and TIME to
#    a linear regression model. 

# OLS regression - mean adjusted Russ

# pos7 == MA Russ
Y_var = RusselDat.iloc[1:,7]
# pos8 == Yesterday Russ
X_var = RusselDat.iloc[1:,8]
# add constant manually (as separate step) in statsmodels. Why???
X_var = sm.add_constant(X_var)
model = sm.OLS(Y_var,X_var)
OLS_meanAdjRussResults = model.fit()
print(OLS_meanAdjRussResults.summary())

```

```python
# OLS regression - mean adjusted Russ

# pos7 == MA Russ
Y_var = RusselDat.iloc[1:,7]
# pos 9 == Half Yesterday Russ
X_var = RusselDat.iloc[1:,0]
# add constant manually (as separate step) in statsmodels. Why???
X_var = sm.add_constant(X_var)
model = sm.OLS(Y_var,X_var)
OLS_HalfRussResults = model.fit()
print(OLS_HalfRussResults.summary())

print('\n'*2 + 'Mean Adjusted Total MSE: %s; Half Total MSE: %s' %(OLS_meanAdjRussResults.mse_total, OLS_HalfRussResults.mse_total) )

```

```python
#scatterplot of russel vs time 

plt.scatter(RusselDat['MA_Russ'],RusselDat['YesterdayRuss'] )
plt.title('Russell Levels')
plt.xlabel('Yesterday Level')
plt.ylabel('Today Level')
```

### Question 3, Returns

```python
# ttest docs: http://www.statsmodels.org/dev/generated/statsmodels.stats.weightstats.ttest_ind.html#statsmodels.stats.weightstats.ttest_ind

meanReturns = np.mean(RusselDat['Returns'])
stdevReturns = np.std(RusselDat['Returns'])
ttestReturns = SPstats.ttest_1samp(RusselDat.iloc[1:,10], 0)
# tstat_manual =  (meanReturns - 0)/(stdevReturns / np.sqrt(len(RusselDat['Returns'])))

print('Mean Ret: ', meanReturns, 'Stdev Ret: ', stdevReturns, 'TStat: ',ttestReturns.statistic, 'Tstat Pvalue: ', ttestReturns.pvalue)
```
