````python
# Solution

import pandas as pd
import numpy as np
import dateutil as du

valid_grades = ['A','N','B','C','Z','P']
inspec_raw = pd.read_csv('DOHMH_New_York_City_Restaurant_Inspection_Results.csv')
initial_rows = inspec_raw.shape[0]
````
