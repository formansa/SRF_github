```python
yogOLS = smr.linear_model.OLS(yogDat_init['YOPLAIT UNITS'], yog_indepVars, hasconst=True)
yogOLS = yogOLS.fit()
yogOLS.summary()
```

```python
ParamsDF = pd.DataFrame(yogOLS.params)
ParamsDF.columns = ['Coef']
ParamsDF.loc['Jan']
```

```python
# Elasticity Calcs 
#yogWpred = pd.concat([yogDat_init, pred_vlaues2['mean']], axis=1)

def OwnPED(df, params):
    # row is a DF
    ownPED_BF = []
    ownPED_AF = []
    #df.reset_index(inplace=True)
    
    #before CHOB
    for ea in df.index[:355]:
        mo = str(yog_indepVars.loc[ea,:][yog_indepVars.loc[ea,:] == 1].index)
        month = mo[8:11]
        row = df.iloc[ea,:]
        newQ =  row['YOPLAIT PRICE']*(params.loc['YOPLAIT PRICE']*1.01) + row['DANNON PRICE']*params.loc['DANNON PRICE']  + row['CHOBANI PRICE']*params.loc['CHOBANI PRICE'] + row[month]*params.loc[month] + row['LogWeek']*params.loc['LogWeek']    
        chngQ = (newQ - yogDat_init.loc[:,'YOPLAIT UNITS'].mean())/ yogDat_init.loc[:,'YOPLAIT UNITS'].mean()
        ownPED = chngQ / .01
        ownPED_BF.append(ownPED)
    #After CHOB   
    for ea in df.index[355:]:
        mo = str(yog_indepVars.loc[ea,:][yog_indepVars.loc[ea,:] == 1].index)
        month = mo[8:11]
        row = df.iloc[ea,:]
        newQ =  row['YOPLAIT PRICE']*(params.loc['YOPLAIT PRICE']*1.01) + row['DANNON PRICE']*params.loc['DANNON PRICE']  + row['CHOBANI PRICE']*params.loc['CHOBANI PRICE'] + row[month]*params.loc[month] + row['LogWeek']*params.loc['LogWeek']    
        chngQ = (newQ - yogDat_init.loc[:,'YOPLAIT UNITS'].mean())/ yogDat_init.loc[:,'YOPLAIT UNITS'].mean()
        ownPED = chngQ / .01
        ownPED_AF.append(ownPED)
    
    return pd.Series(ownPED_BF), pd.Series(ownPED_AF)



def CPED_chobani(df, params):
    # df is a DF, cross prod is a str
    ownPED_BF = []
    ownPED_AF = []
    #df.reset_index(inplace=True)
    
    #After CHOB   
    for ea in df.index[355:]:
        row = df.iloc[ea,:]
        newQ =  row['YOPLAIT PRICE']*params.loc['YOPLAIT PRICE'] + row['DANNON PRICE']*params.loc['DANNON PRICE']  + (row['CHOBANI PRICE']*1.01)*params.loc['CHOBANI PRICE'] + row['MonthNum']*params.loc['MonthNum'] + row['LogWeek']*params.loc['LogWeek']    
        chngQ = (newQ - row['mean'])/ row['mean']
        ownPED = chngQ / .01
        ownPED_AF.append(ownPED)
    
    return 0, pd.Series(ownPED_AF)



def CPED_dannon(df, params):
    # df is a DF, cross prod is a str
    ownPED_BF = []
    ownPED_AF = []
    #df.reset_index(inplace=True)
    
    #before CHOB
    for ea in df.index[:355]:
        row = df.iloc[ea,:]
        newQ =  row['YOPLAIT PRICE']*params.loc['YOPLAIT PRICE'] + (row['DANNON PRICE']*1.01)*params.loc['DANNON PRICE']  + row['CHOBANI PRICE']*params.loc['CHOBANI PRICE'] + row['MonthNum']*params.loc['MonthNum'] + row['LogWeek']*params.loc['LogWeek']    
        chngQ = (newQ - row['mean'])/ row['mean']
        ownPED = chngQ / .01
        ownPED_BF.append(ownPED)
    #After CHOB   
    for ea in df.index[355:]:
        row = df.iloc[ea,:]
        newQ =  row['YOPLAIT PRICE']*params.loc['YOPLAIT PRICE'] + (row['DANNON PRICE']*1.01)*params.loc['DANNON PRICE']  + row['CHOBANI PRICE']*params.loc['CHOBANI PRICE'] + row['MonthNum']*params.loc['MonthNum'] + row['LogWeek']*params.loc['LogWeek']    
        chngQ = (newQ - row['mean'])/ row['mean']
        ownPED = chngQ / .01
        ownPED_AF.append(ownPED)
    
    return pd.Series(ownPED_BF), pd.Series(ownPED_AF)
```

```python
# MAPE Calc

def MAPE_Calc(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


mape = MAPE_Calc(yogDat_init['YOPLAIT UNITS'], pred_vlaues2['mean'])
mape
```
