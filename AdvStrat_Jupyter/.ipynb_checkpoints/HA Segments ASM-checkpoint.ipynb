Relevant Carriers are:
HA, AA, DL, UA, WN

Data from: [USDOT - T-100 Segment Data (US Carriers Only)](https://www.transtats.bts.gov/DataIndex.asp) 

```python
import pandas as pd 
pd.__version__
```

```python
segments2017raw =  pd.read_csv('Segments_dat/2017SegmentsUSCarriers.csv')
HA_segments17 = segments2017raw[segments2017raw['CARRIER'] == 'HA']
```

```python
intl_segs = HA_segments17[(HA_segments17['ORIGIN_STATE_ABR'].isnull() | HA_segments17['DEST_STATE_ABR'].isnull())]
island_segs = HA_segments17[((HA_segments17['ORIGIN_STATE_ABR'] == 'HI') & (HA_segments17['DEST_STATE_ABR'] == 'HI'))]
```

```python
def categorize_mkt(DOT_SEGMENTS_DF):
    DOT_SEGMENTS_DF['Market'] = ''
    
    for indx in DOT_SEGMENTS_DF.index:
        if ((pd.isnull(DOT_SEGMENTS_DF.loc[indx, 'ORIGIN_STATE_ABR'])) or (pd.isnull(DOT_SEGMENTS_DF.loc[indx, 'DEST_STATE_ABR']))):
            DOT_SEGMENTS_DF.loc[indx, 'Market'] = 'INTL'
        if ((DOT_SEGMENTS_DF.loc[indx,'ORIGIN_STATE_ABR'] == 'HI') & (DOT_SEGMENTS_DF.loc[indx,'DEST_STATE_ABR'] == 'HI')):
            DOT_SEGMENTS_DF.loc[indx, 'Market'] = 'ISLAND'
        else:
            DOT_SEGMENTS_DF.loc[indx, 'Market'] = 'DOM'
            
    
    return DOT_SEGMENTS_DF
```

```python
func_categorize = categorize_mkt(HA_segments17)
```

```python
func_categorize.loc[list(intl_segs.index),'Market'] = 'INTL'
```

```python
func_categorize.loc[:,['DEST_CITY_NAME', 'ORIGIN_CITY_NAME', 'Market']]
```

```python
func_categorize['ASM'] = func_categorize['SEATS'] * func_categorize['DISTANCE']
HA_segmentsGRP17 = func_categorize.groupby('Market')
```

```python
ASM_byMkt = HA_segmentsGRP17['ASM'].sum()
Seat_byMkt = HA_segmentsGRP17['SEATS'].sum()
ASM_byMkt = pd.DataFrame(ASM_byMkt)
Seat_byMkt = pd.DataFrame(Seat_byMkt)

```

```python
mkt_stats = pd.concat([ASM_byMkt, Seat_byMkt], axis=1)
mkt_stats['Pct_ASM'] = mkt_stats['ASM'] / mkt_stats['ASM'].sum()
mkt_stats['Pct_SEATS'] = mkt_stats['SEATS'] / mkt_stats['SEATS'].sum()
```
