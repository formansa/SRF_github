### Functions to Repeat for All Yrs

```python

def summarizeSegData(year):
    yr = str(year)
    
    segmentsRaw = pd.read_csv('Segments_dat/' + yr + 'SegmentsUSCarriers.csv')
    
    #narrow DF to records of flights with orig or dest in HI or records with BOTH orig and dest in HI
    HIwTF = segmentsRaw[(segmentsRaw['DEST_STATE_NM'] == 'Hawaii')| (segmentsRaw['ORIGIN_STATE_NM'] == 'Hawaii')]
    HI_intra = segmentsRaw[(segmentsRaw['DEST_STATE_NM'] == 'Hawaii') & (segmentsRaw['ORIGIN_STATE_NM'] == 'Hawaii')]
    
    #group dataframes by carrier name
    TF_HI = HIwTF.groupby('UNIQUE_CARRIER_NAME')
    intra_HI = HI_intra.groupby('UNIQUE_CARRIER_NAME')
    
    
    # count flights
    inter_CntFlights = TF_HI['UNIQUE_CARRIER'].count()
    inter_CntFlights.rename('CountFlights Inter', inplace=True)
    intra_CntFlights = intra_HI['UNIQUE_CARRIER'].count()
    intra_CntFlights.rename('CountFlights Intra', inplace=True)
    #sum seats
    inter_SumSeats = TF_HI['SEATS'].sum()
    inter_SumSeats.rename('SumSeats Inter', inplace=True)
    intra_SumSeats = intra_HI['SEATS'].sum()
    intra_SumSeats.rename('SumSeats Intra', inplace=True)
    #sum passangers
    inter_SumPass = TF_HI['PASSENGERS'].sum()
    inter_SumPass.rename('SumPass Inter', inplace=True)
    intra_SumPass = intra_HI['PASSENGERS'].sum()
    intra_SumPass.rename('SumPass Intra', inplace=True)

    summary_series = [inter_CntFlights, intra_CntFlights, inter_SumSeats, intra_SumSeats, inter_SumPass, intra_SumPass]

    smryDF = pd.concat(summary_series, axis=1)

    smryDF['PctPass Inter'] = smryDF['SumPass Inter'] / smryDF['SumPass Inter'].sum() * 100
    smryDF['PctPass Intra'] = smryDF['SumPass Intra'] / smryDF['SumPass Intra'].sum() * 100
    
    smryDF['Year'] = yr
    smryDF.set_index('Year',drop=True, append=True, inplace=True)

    return smryDF

```

```python
seg_yrs = list(range(2013,2018))
df_li = []

for ea in seg_yrs:
    summedDF = summarizeSegData(ea)
    df_li.append(summedDF)
    
    
full_DF = pd.concat(df_li)
    
```

```python
full_DF.to_csv('Segments_dat/SegmentRollup.csv')
```
